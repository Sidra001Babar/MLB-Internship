# Text Preprocessing


## Step 1: Import Required Libraries
### Explanation
- **re**: Used for cleaning text with regular expressions.
- **nltk**: Natural Language Toolkit for text processing.
- **stopwords**: Provides common English words (e.g., *the, is, and*) that are usually removed.
- **WordNetLemmatizer**: Converts words to their base (dictionary) form.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

## Step 2: Download Required NLTK Resources

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Step 3: Initialize Stopwords and Lemmatizer
### Explanation
- Loads English stopwords into a set for fast lookup.
- Creates a lemmatizer object that will reduce words to their root form.

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

# Text Preprocessing Function

```python
def preprocess(text):

    print("=" * 60)
    print("Original Text:")
    print(text)

    # 1. Lowercase
    text = text.lower()
    print("\n1. After Lowercase:")
    print(text)

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    print("\n2. After Removing URLs:")
    print(text)

    # 3. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    print("\n3. After Removing Punctuation:")
    print(text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    print("\n4. After Removing Numbers:")
    print(text)

    # 5. Tokenization
    tokens = text.split()
    print("\n5. After Tokenization:")
    print(tokens)

    # 6. Stopword Removal
    tokens = [word for word in tokens if word not in stop_words]
    print("\n6. After Stopword Removal:")
    print(tokens)

    # 7. Lemmatization:
    tokens = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
    print("\n7. After Lemmatization:")
    print(tokens)

    # 8. Join Tokens
    cleaned_text = " ".join(tokens)
    print("\n8. Final Cleaned Text:")
    print(cleaned_text)

    print("=" * 60)

    return cleaned_text
```

```python
raw_text = """
Hello Everyone!! My name is Sidra. I have 2 cats and 3 dogs.
I am learning NLP in 2026.
Visit https://www.example.com for more information.
The cats are running, jumping, and playing happily!!!
"""

# Apply preprocessing
cleaned_text = preprocess(raw_text)
print("\nCleaned Text:")
print(cleaned_text)
```

# Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2)
)

# Since TF-IDF expects a collection of documents, pass cleaned_text as a list
X_tfidf = tfidf.fit_transform([cleaned_text])

# Display feature names
print("Features:")
print(tfidf.get_feature_names_out())

# Display TF-IDF matrix
print("\nTF-IDF Matrix:")
print(X_tfidf.toarray())
```

## Load Datatset

```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
```

# Basic Information

```python
print("Basic Info")
print("\nFirst five records")
print(df.head())

print("\nLast five records:")
print(df.tail())

print("\nColumn Names:")
print(df.columns)

print("\nDataset Infor")
df.info()

df.describe()
```

# Missing vLues

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

# Average of each Subject

```python
print("\nAverage Marks")

print("Math Average:", df["math score"].mean())
print("Reading Average:", df["reading score"].mean())
print("Writing Average:", df["writing score"].mean())

# Identify the top 5 performing students
df["Total Marks"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
)
```

# Top Five performing

```python
df = df.sort_values(
    by="Total Marks",
    ascending=False
)
top5 = df.head(5)
print(top5)
```

# Below Average

```python
average_total = df["Total Marks"].mean()
below_average = df[df["Total Marks"] < average_total]
print(below_average)
```

# Total Students

```python
total_students = len(df)
print(total_students)
```

# Save Analyzed data into new file

```python
output_file = "StudentsPerformance_Analyzed.csv"

df.to_csv(
    output_file,
    index=False
)

print("\nAnalyzed dataset saved successfully.")
```

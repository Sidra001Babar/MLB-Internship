# Python Lists

## What is a List?

A **list** is a built-in Python data type used to store **multiple items in a single variable**.

Lists are one of Python's **four built-in collection data types**:

- **List** → Ordered, changeable, allows duplicates.
- **Tuple** → Ordered, unchangeable, allows duplicates.
- **Set** → Unordered, changeable,unchangeable*(frozenset), no duplicates.
- **Dictionary** → Ordered**, changeable, stores key-value pairs.

---
# Creating a List

Lists are created using **square brackets `[]`**.

### Syntax

````python
mylist = ["apple", "banana", "cherry"]
print(mylist)
````

list can also be created by using the `list()` constructor.

````python
thislist = list(("apple", "banana", "cherry"))
print(thislist)
````

> **Note:** When using `list()`, notice the **double parentheses**:
>
> - Outer parentheses belong to `list()`
> - Inner parentheses create a tuple that is converted into a list.

# List Indexing

Every item in a list has an index.

Indexes start from **0**.

| Index | Value |
|-------:|-------|
| 0 | apple |
| 1 | banana |
| 2 | cherry |

Example

````python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
````

# List Length

Use the built-in `len()` function to find the number of items in a list.

Example

````python
fruits = ["apple", "banana", "cherry"]

print(len(fruits))
````

# Python - Access List Items - By Positive Indexing, Negative indexing, Slicing

## Accessing List Items

Lists are **indexed**, which means every item has a position called an **index**.

The first item always starts at **index `0`**.

### Example

````python
thislist = ["apple", "banana", "cherry"]

print(thislist[1])
````

### Index Positions

| Index | Value |
|------:|-------|
| 0 | apple |
| 1 | banana |
| 2 | cherry |


> **Note:** Python indexing starts from **0**, not 1.

# Positive Indexing

Positive indexes count from **left to right**.

```text
Index:    0        1         2
        ----------------------------
List :  apple   banana   cherry
```

Example

````python
fruits = ["apple", "banana", "cherry"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
````

# Negative Indexing

Negative indexing starts counting from the **end of the list**.

| Negative Index | Value |
|--------------:|-------|
| -1 | cherry |
| -2 | banana |
| -3 | apple |

### Visualization

```text
Positive Index:   0        1         2
                 ------------------------
List:           apple   banana   cherry
                 ------------------------
Negative Index: -3       -2       -1
```

````python
thislist = ["apple", "banana", "cherry"]

print(thislist[-1])
````

````python
print(thislist[-2])
````

# Range of Indexes (List Slicing)

Instead of accessing one item, you can access **multiple items** using **slicing**.
means [start index:end index - 1] [start: end] : in this case sliicing start from start index and ends at end index

````python
thislist = [
    "apple",
    "banana",
    "cherry",
    "orange",
    "kiwi",
    "melon",
    "mango"
]

print(thislist[2:5])
````

### Explanation

```text
Index :   0       1       2        3       4       5       6
         -------------------------------------------------------
List :  apple  banana  cherry  orange   kiwi   melon   mango
                     ↑--------------------↑
                    start=2             end=5

````python
print(thislist[:4])
````

# Omitting the End Index

If the **end index is omitted**, Python returns items until the end of the list.

````python
print(thislist[2:])
````

````python
print(thislist[:]) # return the complete list
````

````python
print(thislist[-4:-1])
````

````python
print(thislist[-1:-4])  # can be move backward so always empty
````

# Python List Methods

Python provides several built-in methods to perform operations on lists. Below is an explanation of the most commonly used list methods.

## 1. `append()`

**Purpose:**  
Adds a single element to the **end** of the list.


````python
fruitss = ["Apple", "Banana"]
fruitss.append("Orange")
print(fruitss)
````

````python
otherlist = ["banana juice","apple juice"]
fruitss.append(otherlist)
print(fruitss)
````

## 2. `count()`

**Purpose:**  
Returns the number of times a specified element appears in the list.


````python
numbers = [1, 2, 2, 3, 2]
print(numbers.count(2))
````

## 3. `extend()`

**Purpose:**  
Adds all elements from another list (or any iterable) to the end of the current list.


````python
list1 = [1, 2]
list2 = [3, 4]
list1.extend(list2)
print(list1)
````

## 4. `index()`

**Purpose:**  
Returns the index (position) of the **first occurrence** of the specified element.

````python
fruits = ["Apple", "Banana", "Orange"]
print(fruits.index("Banana"))
````

## 5. `insert()`

**Purpose:**  
Inserts an element at a specified index.
```python
list_name.insert(index, item)
```

````python
numbers = [1, 3, 4]
numbers.insert(1, 2)
print(numbers)
````

## 6. `pop()`

**Purpose:**  
Removes and returns the element at the specified index. If no index is given, it removes the last element.


````python
colors = ["Red", "Green", "Blue"]
removed = colors.pop(1)
print(removed)
print(colors)
````

## 7. `copy()`
**Purpose:**  
Creates and returns a **shallow copy** of the list.

````python
list1 = [1, 2, 3]
list2 = list1.copy()
print(list2)
list1.append(4)
print(list2)
````

````python
# when not use shallow copy function, change in original list will also reflect in copied
list1 = [1, 2, 3]
list2 = list1
print(list2)
list1.append(4)
print(list2)
````


## 8. `remove()`

**Purpose:**  
Removes the **first occurrence** of the specified element.

````python
fruits = ["Apple", "Banana", "Orange"]
fruits.remove("Banana")
print(fruits)
````

## 9. `reverse()`

**Purpose:**  
Reverses the order of elements in the list.

````python
numbers = [1, 2, 3, 4]
numbers.reverse()
print(numbers)
````

## 10. `sort()`

**Purpose:**  
Sorts the list in **ascending order** by default.

````python
numbers = [5, 2, 8, 1]
numbers.sort()
print(numbers)
````

**Descending Order:**

````python
numbers.sort(reverse=True)
print(numbers)
````

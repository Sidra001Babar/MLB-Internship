# Tuples

## Explanation
- Like lists, tuples are used to store multiple items in a single variable.
- A tuple is an **ordered** collection, which means the items keep their order.
- Tuples are **unchangeable (immutable)**, meaning their values cannot be modified after creation.
- Tuples are written using **round brackets ()**.
- Each item has an index, starting from **0**.

## Example
The tuple contains three fruit names and prints all of them.

## Output
('apple', 'banana', 'cherry')

```python
thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

# Difference Between List and Tuple

| Feature | List | Tuple |
|---------|------|-------|
| Mutable | Yes | No |
| Iteration | Slower | Faster |
| Best Used For | Insertion and deletion | Read-only operations |
| Memory | Uses more memory | Uses less memory |
| Methods | More built-in methods | Fewer built-in methods |
| Safety | More error-prone | Safer because it cannot be changed |

## Explanation
- Lists allow changes after creation.
- Tuples cannot be modified once created.
- Tuples are faster and use less memory.

```python
my_list = [10, 20, 30, 40]
my_tuple = (10, 20, 30, 40)

print(my_list)
print(my_tuple)
```

# Create a Tuple with One Item

## Explanation
- To create a tuple containing only one item, place a comma after the item.
- Without the comma, Python treats it as a normal string instead of a tuple.

## Output

<class 'tuple'>
<class 'str'>

```python
thistuple = ("apple",)
print(type(thistuple))

# NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

# Change Tuple Values

## Explanation
- Tuples are immutable, so their values cannot be changed directly.
- A workaround is:
  1. Convert the tuple into a list.
  2. Change the required value.
  3. Convert the list back into a tuple.



```python
x = ("apple", "banana", "cherry")

x[1] = "gava"
```

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
```

# Add Items to a Tuple

## Explanation
- Tuples do not have an `append()` method because they are immutable.
- To add an item:
  1. Convert the tuple into a list.
  2. Use `append()`.
  3. Convert the list back into a tuple.



```python
thistuple = ("apple", "banana", "cherry")
thistuple.append("orange")
```

```python

y = list(thistuple)

y.append("orange")

thistuple = tuple(y)

print(thistuple)
```

# Add Items to a Tuple

## Explanation
- Tuples are immutable, so they cannot be modified directly.
- Another way to add an item is to create a new tuple containing the new item and add it to the existing tuple using the `+` operator.

## Syntax

tuple_name += (new_item,)

## Example
A new tuple containing `"orange"` is created and added to the existing tuple.



```python
thistuple = ("apple", "banana", "cherry")

y = ("orange",)

thistuple += y

print(thistuple)
```

# Remove Items from a Tuple

## Explanation
- Tuples are immutable, so items cannot be removed directly.
- To remove an item:
  1. Convert the tuple into a list.
  2. Use the `remove()` method.
  3. Convert the list back into a tuple.



```python
thistuple = ("apple", "banana", "cherry")

y = list(thistuple)

y.remove("apple")

thistuple = tuple(y)

print(thistuple)
```

# Delete a Tuple

## Explanation
- The `del` keyword deletes the entire tuple.
- After deleting the tuple, it no longer exists.
- Trying to access it raises a `NameError`.

## Output

NameError: name 'thistuple' is not defined

```python
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple)
```

# Packing and Unpacking a Tuple

## Packing

Packing means storing multiple values inside a tuple.

### Example

fruits = ("apple", "banana", "cherry")

## Unpacking

Unpacking means assigning tuple values to separate variables.

The number of variables should match the number of tuple items.



```python
fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

# Using Asterisk (*)

## Explanation
- If there are fewer variables than tuple values, use an asterisk (`*`) before one variable.
- The remaining values are stored as a list in that variable.



```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

# Loop Through a Tuple

## Explanation
- A `for` loop is used to access each item in a tuple one by one.
- Each iteration prints one element of the tuple.



```python
thistuple = ("apple", "banana", "cherry")

for x in thistuple:
    print(x)
```

# Join Two Tuples

## Explanation
- Two or more tuples can be joined using the `+` operator.
- A new tuple containing the elements of both tuples is created.

## Output

('a', 'b', 'c', 1, 2, 3)

```python
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2

print(tuple3)
```

# Multiply Tuples

## Explanation
- The `*` operator repeats the contents of a tuple a specified number of times.
- It creates a new tuple with repeated elements.

## Syntax

tuple_name * number

## Output

('apple', 'banana', 'cherry', 'apple', 'banana', 'cherry')

```python
fruits = ("apple", "banana", "cherry")

mytuple = fruits * 2

print(mytuple)
```

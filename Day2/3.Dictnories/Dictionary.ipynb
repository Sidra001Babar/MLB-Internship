# What are Dictionaries?

## Explanation
- A dictionary is a data structure that stores data in the form of **key-value pairs**.
- Each key is associated with a specific value.
- Dictionaries are:
  - **Unordered** (items are stored as key-value pairs and are not accessed by index).
  - **Mutable**, meaning values can be changed after creation.
  - **Dynamic**, meaning new key-value pairs can be added or removed.
- Dictionaries are also known as:
  - Associative Arrays
  - Hash Maps (in other programming languages)

## Example

Key      → Value

name     → John Smith

age      → 20

major    → Computer Science

```python
student = {
    "name": "John Smith",
    "age": 20,
    "major": "Computer Science"
}

print(student)
```

# Dictionary Basics

## Explanation
- Dictionaries are created using **curly braces `{}`**.
- Each item consists of:
  - A **key**
  - A **value**
- The key and value are separated by a colon (`:`).
- Each key must be:
  - **Unique** (no duplicate keys).
  - **Immutable** (cannot be changed, e.g., strings, integers, tuples).

## Syntax

dictionary = {
    key1: value1,
    key2: value2
}

# Empty Dictionary

## Explanation
An empty dictionary can be created in two ways:

1. Using curly braces `{}`.
2. Using the `dict()` constructor.

Both methods create an empty dictionary.

```python
dict1 = {}

dict2 = dict()

print(dict1)
print(dict2)
```

# Using dict() Constructor

## Explanation
The `dict()` constructor creates a dictionary by passing key-value pairs as arguments.

## Syntax

dict(key=value)



```python
dict3 = dict(name="John", age=20)

print(dict3)
```

# Create Dictionary from List of Tuples

## Explanation
A dictionary can also be created from a list containing tuples.
Each tuple contains:
- A key
- A value

The `dict()` constructor converts the list into a dictionary.

```python
pairs = [("apple", 1), ("banana", 2)]

dict4 = dict(pairs)

print(dict4)
```

# Accessing Values Using Square Brackets

## Explanation
- Dictionary values can be accessed using their key inside square brackets.
- If the key exists, its corresponding value is returned.

## Syntax

dictionary[key]


```python
student = {
    "name": "John",
    "age": 20
}

print(student["name"])
```

# Accessing Values Using get()

## Explanation
- The `get()` method returns the value of the specified key.
- It is safer than square brackets because it allows a default value to be returned if the key does not exist.

## Syntax

dictionary.get(key)

dictionary.get(key, default_value)

## Example

If the key is not found, `"N/A"` is returned.

```python
student = {
    "name": "John",
    "age": 20
}

print(student.get("age"))

print(student.get("grade", "N/A"))
```

# Adding a New Key-Value Pair

## Explanation
A new key-value pair can be added by assigning a value to a new key.

## Syntax

dictionary[new_key] = value


```python
student = {
    "name": "John",
    "age": 20
}

student["grade"] = "A"

print(student)
```

# Updating an Existing Value

## Explanation
If the key already exists, assigning a new value updates that value.

## Syntax

dictionary[key] = new_value

```python
student = {
    "name": "John",
    "age": 20
}

student["age"] = 21

print(student)
```

# Updating Multiple Values

## Explanation
The `update()` method updates one or more key-value pairs at the same time.

## Syntax

dictionary.update({key:value})

```python
student = {
    "name": "John",
    "age": 21
}

student.update({
    "age": 22,
    "major": "Physics"
})

print(student)
```

# Remove a Specific Key

## Explanation
The `del` keyword removes a specified key-value pair from the dictionary.

## Syntax

del dictionary[key]

```python
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

del student["grade"]

print(student)
```

# pop() Method

## Explanation
- The `pop()` method removes the specified key.
- It returns the removed value.
- In this example, the removed value is stored with a new key named `score`.

## Syntax

dictionary.pop(key)

```python
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

student["score"] = student.pop("grade")

print(student)
```

# popitem() Method

## Explanation
- The `popitem()` method removes and returns the last inserted key-value pair from the dictionary.

## Syntax

dictionary.popitem()

```python
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

last_item = student.popitem()

print(last_item)
print(student)
```

# clear() Method

## Explanation
The `clear()` method removes all key-value pairs from the dictionary, making it empty.

## Syntax

dictionary.clear()

```python
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

student.clear()

print(student)
```

# keys() Method

## Explanation
- The `keys()` method returns a view object containing all the keys in the dictionary.
- It is useful when you want to access only the keys.

## Syntax

dictionary.keys()



```python
student = {
    "name": "John",
    "age": 20
}

print(student.keys())
```

# values() Method

## Explanation
- The `values()` method returns a view object containing all the values in the dictionary.

## Syntax

dictionary.values()



```python
student = {
    "name": "John",
    "age": 20
}

print(student.values())
```

# items() Method

## Explanation
- The `items()` method returns all key-value pairs as tuples.

## Syntax

dictionary.items()

## Output

dict_items([('name', 'John'), ('age', 20)])

```python
student = {
    "name": "John",
    "age": 20
}

print(student.items())
```

# Dictionary Comprehension

## Explanation
- Dictionary comprehension provides a short way to create dictionaries.
- It follows the syntax:

{key_expression : value_expression for item in iterable}

## Example

Create a dictionary containing numbers and their squares.


```python
squares = {x: x**2 for x in range(5)}

print(squares)
```

# Dictionary Comprehension with Condition

## Explanation
- An `if` condition can be used to include only selected items.
- In this example, only even numbers are included.


```python
even_squares = {
    x: x**2
    for x in range(5)
    if x % 2 == 0
}

print(even_squares)
```

# Nested Dictionaries

## Explanation
- A dictionary can contain another dictionary as its value.
- Such dictionaries are called **nested dictionaries**.
- Nested dictionaries help organize related information.

## Example
Each student has:
- Age
- Grades
- Another dictionary containing subject grades.

```python
students = {
    "John": {
        "age": 20,
        "grades": {
            "math": "A",
            "physics": "B"
        }
    },
    "Mary": {
        "age": 19,
        "grades": {
            "math": "B",
            "physics": "A"
        }
    }
}

print(students)
```

# Dictionary Performance

## Explanation
- Dictionaries use a **hash table** implementation.
- This allows operations like searching, inserting, and deleting to be very fast.

### Average Time Complexity

| Operation | Complexity |
|-----------|------------|
| Search | O(1) |
| Insert | O(1) |
| Delete | O(1) |

### Comparison with Lists

| List | Dictionary |
|------|------------|
| Search: O(n) | Search: O(1) |

Because dictionaries search by key, they are generally much faster than searching through a list.

# Memory Usage

## Explanation
- Dictionaries use more memory than lists.
- Each key-value pair stores:
  - The hash of the key.
  - The key itself.
  - The value.
  - Additional metadata required by the hash table.

### Summary

- Lists use less memory.
- Dictionaries use more memory but provide faster searching.

# Best Practices

## Explanation

When working with dictionaries, follow these good programming practices:

### 1. Use Meaningful Keys
Choose keys that clearly describe the stored value.

Example:
- `"name"` instead of `"n"`
- `"student_id"` instead of `"id"`

### 2. Check for Key Existence
Before accessing a key, ensure it exists to avoid errors.

### 3. Use get() with Default Values
The `get()` method returns a default value if the key is not found.

### 4. Consider Using collections.defaultdict
`defaultdict` automatically creates default values for missing keys.

### 5. Keep Keys Immutable
Dictionary keys should be immutable data types such as:
- Strings
- Integers
- Tuples

Avoid using mutable objects like lists as keys.

# Special Dictionary Types

The `collections` module provides specialized dictionary types.

These include:

- defaultdict
- OrderedDict
- Counter

```python
from collections import defaultdict, OrderedDict, Counter
```

# defaultdict

## Explanation

A `defaultdict` automatically creates a default value for missing keys.

This prevents a `KeyError`.

In this example, a new list is automatically created for `"new_key"`.

## Output

defaultdict(<class 'list'>, {'new_key': [1]})

```python
from collections import defaultdict

d = defaultdict(list)

d["new_key"].append(1)

print(d)
```

# OrderedDict

## Explanation

`OrderedDict` maintains the insertion order of dictionary items.

It was mainly used before Python 3.7.

## Example

```python
from collections import OrderedDict

od = OrderedDict()

od["A"] = 1
od["B"] = 2

print(od)
```

# Counter

## Explanation

`Counter` counts how many times each element appears in a collection.

## Output

Counter({'a': 3, 'b': 1, 'c': 1})

```python
from collections import Counter

c = Counter(['a', 'b', 'a', 'c', 'a'])

print(c)
```

# Dictionary Views

## Explanation

Dictionary views provide dynamic views of:

- Keys
- Values
- Key-value pairs

If the dictionary changes, the views are automatically updated.

Methods:

- keys()
- values()
- items()

```python
d = {
    "a": 1,
    "b": 2
}

keys = d.keys()
values = d.values()
items = d.items()

d["c"] = 3

print(keys)
print(values)
print(items)
```

# Type Hints with Dictionaries

## Explanation

Type hints specify the expected data types of dictionary keys and values.

They improve code readability and help with static type checking.

The `typing` module provides:

- Dict
- List
- Union

```python
from typing import Dict, List, Union

grades: Dict[str, float] = {
    "math": 95.5,
    "physics": 89.0
}

student: Dict[str, Union[str, int, List[str]]] = {
    "name": "John",
    "age": 20,
    "courses": ["math", "physics"]
}

print(grades)
print(student)
```

# Set-like Operations

## Explanation

Dictionary keys behave like sets.

The following operations can be performed:

- Union (`|`)
- Intersection (`&`)
- Difference (`-`)

```python
d1 = {
    "a": 1,
    "b": 2
}

d2 = {
    "b": 3,
    "c": 4
}

keys1 = d1.keys()
keys2 = d2.keys()

print(keys1 | keys2)

print(keys1 & keys2)

print(keys1 - keys2)
```

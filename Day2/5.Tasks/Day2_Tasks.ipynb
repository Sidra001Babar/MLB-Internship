# 1.List

Find the largest number in a list.

```python
myList = [1, 2, 3, 9, 7, 8]

largest = myList[0]
for num in myList:
    if num > largest:
        largest = num
print("Largest number:", largest)
```

Find the second largest number.

```python
myList = [1, 2, 3, 9, 7, 8]

largest = myList[0]
secLargest = myList[0]

for num in myList:
    if num > largest:
        secLargest = largest
        largest = num
    elif num > secLargest and num != largest:
        secLargest = num

print("Largest number:", largest)
print("Second Largest number:", secLargest)
```

Remove duplicate values from a list.

```python
myList = [1, 2, 3, 2, 4, 1, 5, 3, 6]

uniqueList = []

for item in myList:
    if uniqueList.count(item) == 0:
        uniqueList.append(item)

print("Original List:", myList)
print("List after removing duplicates:", uniqueList)
```

Reverse a list without using built-in reverse().

```python
myList = [1, 2, 3, 4, 5]

reversedList = []

for i in range(-1, -len(myList)-1, -1):
    reversedList.append(myList[i])

print(reversedList)
```

Find common elements between two lists.

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common = []

for item in list1:
    if list2.count(item) > 0:
        common.append(item)

print("Common elements:", common)
```

# Set

Perform union and intersection operations.

```python
# UNION

set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

print("union():", set1.union(set2))
print("| operator:", set1 | set2)

# Union with multiple sets
set3 = {"John", "Elena"}
set4 = {"apple", "banana", "cherry"}

print("Multiple union:", set1.union(set2, set3, set4))
print("Multiple |:", set1 | set2 | set3 | set4)

# Union with a tuple
x = {"a", "b", "c"}
y = (1, 2, 3)

print("Union with tuple:", x.union(y))

# UPDATE

set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set1.update(set2)
print("update():", set1)


# INTERSECTION

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

print("intersection():", set1.intersection(set2))
print("& operator:", set1 & set2)
```

# Dictionaries

Create a student record dictionary.

```python
students = {
    "student1": {
        "name": "Ali",
        "courses": {
            "Python": 90,
            "Database": 85
        }
    },
    "student2": {
        "name": "Sara",
        "courses": {
            "Python": 95,
            "Database": 80
        }
    }
}

print(students)
```

Calculate average marks of students.

```python
python_avg = (
    students["student1"]["courses"]["Python"] +
    students["student2"]["courses"]["Python"]
) / 2

database_avg = (
    students["student1"]["courses"]["Database"] +
    students["student2"]["courses"]["Database"]
) / 2

print("Python Average:", python_avg)
print("Database Average:", database_avg)
```

Count frequency of words in a sentence.

```python
sentence = "python is easy and python is fun"

words = sentence.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)
```

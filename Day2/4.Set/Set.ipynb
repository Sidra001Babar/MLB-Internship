# What is a Set?

## Explanation
- A **Set** is a built-in Python data type used to store multiple items in a single variable.
- It is one of Python's four collection data types:
  - List
  - Tuple
  - Set
  - Dictionary
- A set is:
  - **Unordered** – Items do not have a fixed order.
  - **Unchangeable** – Individual items cannot be changed after creation.
  - **Unindexed** – Items cannot be accessed using an index.
- Although set items cannot be changed individually, you can:
  - Add new items.
  - Remove existing items.
- Sets are written using **curly braces `{}`**.

## Syntax

set_name = {item1, item2, item3}



````python
thisset = {"apple", "banana", "cherry"}

print(thisset)
````

# True and 1

## Explanation
- In Python sets, `True` and `1` are treated as the same value.
- Therefore, only one of them is stored in the set.



````python
thisset = {"apple", "banana", "cherry", True, 1, 2}

print(thisset)
````

# False and 0

## Explanation
- In Python sets, `False` and `0` are considered equal.
- Only one of these values is stored in the set.



````python
thisset = {"apple", "banana", "cherry", False, True, 0}

print(thisset)  # notice there is no order
````

# Accessing Set Items

## Explanation
- Sets are **unordered**, so they do not have indexes.
- Therefore, you **cannot access** set items using an index or a key.
- To access the items in a set, you can:
  - Loop through the set using a `for` loop.
  - Check whether an item exists using the `in` keyword.
  - Check whether an item does not exist using the `not in` keyword.

**Note:** Since sets are unordered, the order of items printed may vary each time the program runs.

````python
thisset = {"apple", "banana", "cherry"}

for x in thisset:
    print(x)
````

# Adding Items to a Set

## Explanation

Sets are **unchangeable**, which means you cannot modify an existing item. However, you **can add new items** to a set.

There are two common ways to add items:

### 1. `add()` Method
- The `add()` method is used to add **one item** to a set.

**Syntax**

```python
set_name.add(item)
```

### 2. `update()` Method
- The `update()` method is used to add **multiple items** to a set.
- The items can come from:
  - Another set
  - A list
  - A tuple
  - A dictionary
  - Any other iterable object

**Syntax**

```python
set_name.update(iterable)
```
**Note:** Since sets do not allow duplicate values, any duplicate items are ignored automatically.

````python
# Using add() to add one item
thisset = {"apple", "banana", "cherry"}

thisset.add("orange")

print("After add():", thisset)


# Using update() with another set
tropical = {"pineapple", "mango", "papaya"}

thisset.update(tropical)

print("After update() with a set:", thisset)


# Using update() with a list
mylist = ["kiwi", "orange"]

thisset.update(mylist)

print("After update() with a list:", thisset)
````

# Removing Items from a Set

## Explanation

Python provides several methods to remove items from a set.

### 1. `remove()` Method
- Removes the specified item from the set.
- If the item does **not exist**, a **KeyError** is raised.

**Syntax**

```python
set_name.remove(item)
```

### 2. `discard()` Method
- Removes the specified item from the set.
- If the item does **not exist**, **no error** is raised.

**Syntax**

```python
set_name.discard(item)
```

### 3. `pop()` Method
- Removes and returns a **random item** from the set.
- Since sets are unordered, you cannot predict which item will be removed.

**Syntax**

```python
removed_item = set_name.pop()
```

### 4. `clear()` Method
- Removes **all items** from the set.
- The set still exists but becomes empty.

**Syntax**

```python
set_name.clear()
```

### 5. `del` Keyword
- Deletes the **entire set**.
- After deletion, the set no longer exists.
- Accessing the set afterward raises a **NameError**.

**Syntax**

```python
del set_name
```


| Method | Purpose | If Item Does Not Exist |
|---------|---------|------------------------|
| `remove()` | Removes a specified item | Raises **KeyError** |
| `discard()` | Removes a specified item | No error |
| `pop()` | Removes and returns a random item | Removes one random item |
| `clear()` | Removes all items | Set becomes empty |
| `del` | Deletes the entire set | Set no longer exists |

````python
# Original set
thisset = {"apple", "banana", "cherry"}

# 1. remove()
thisset.remove("banana")
print("After remove():", thisset)

# 2. discard()
thisset.discard("apple")
print("After discard():", thisset)

# 3. pop()
removed_item = thisset.pop()
print("Removed item using pop():", removed_item)
print("After pop():", thisset)

# 4. clear()
thisset = {"apple", "banana", "cherry"}
thisset.clear()
print("After clear():", thisset)

# 5. del
thisset = {"apple", "banana", "cherry"}
del thisset

# Uncommenting the next line will raise a NameError
# print(thisset)
````

# Joining Sets in Python

## Explanation

Python provides several methods and operators to join or compare sets.

### 1. `union()`
- Combines all unique elements from two or more sets.
- Returns a **new set**.
- Duplicate values are removed automatically.

**Syntax**

```python
new_set = set1.union(set2)
```

**Operator**

```python
new_set = set1 | set2
```

> **Note:** The `|` operator works only with sets, while `union()` can also combine a set with other iterables such as tuples and lists.

---

### 2. `update()`
- Adds all elements from another set (or iterable) into the original set.
- Modifies the original set.
- Does not return a new set.

**Syntax**

```python
set1.update(set2)
```

---

### 3. `intersection()`
- Returns a **new set** containing only the common (duplicate) elements present in both sets.

**Syntax**

```python
new_set = set1.intersection(set2)
```

**Operator**

```python
new_set = set1 & set2
```

---

### 4. `intersection_update()`
- Keeps only the common elements.
- Modifies the original set instead of creating a new one.

**Syntax**

```python
set1.intersection_update(set2)
```

---

### 5. `difference()`
- Returns a **new set** containing elements that exist only in the first set.

**Syntax**

```python
new_set = set1.difference(set2)
```

**Operator**

```python
new_set = set1 - set2
```

---

### 6. `difference_update()`
- Removes the common elements from the first set.
- Modifies the original set.

**Syntax**

```python
set1.difference_update(set2)
```

---

### 7. `symmetric_difference()`
- Returns a **new set** containing elements that are present in either set, but **not in both**.

**Syntax**

```python
new_set = set1.symmetric_difference(set2)
```

**Operator**

```python
new_set = set1 ^ set2
```

---

### 8. `symmetric_difference_update()`
- Keeps only the non-common elements.
- Modifies the original set.

**Syntax**

```python
set1.symmetric_difference_update(set2)
```

---


**Notes**
- `union()` can join a set with other iterables (such as lists and tuples), while the `|` operator only works with sets.
- `&`, `-`, and `^` operators also work only with sets.
- Duplicate values are automatically removed because sets store only unique elements.

````python

# UNION

set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

print("union():", set1.union(set2))
print("| operator:", set1 | set2)

# Union with multiple sets
set3 = {"John", "Elena"}
set4 = {"apple", "banana", "cherry"}

print("Multiple union:", set1.union(set2, set3, set4))
print("Multiple |:", set1 | set2 | set3 | set4)

# Union with a tuple
x = {"a", "b", "c"}
y = (1, 2, 3)

print("Union with tuple:", x.union(y))

# UPDATE

set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set1.update(set2)
print("update():", set1)


# INTERSECTION

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

print("intersection():", set1.intersection(set2))
print("& operator:", set1 & set2)

# intersection_update()
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.intersection_update(set2)
print("intersection_update():", set1)

# True, False, 1, 0 example
set1 = {"apple", 1, "banana", 0, "cherry"}
set2 = {False, "google", 1, "apple", 2, True}

print("True/False duplicates:", set1.intersection(set2))

# DIFFERENCE

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

print("difference():", set1.difference(set2))
print("- operator:", set1 - set2)

# difference_update()
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.difference_update(set2)
print("difference_update():", set1)

# SYMMETRIC DIFFERENCE

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

print("symmetric_difference():", set1.symmetric_difference(set2))
print("^ operator:", set1 ^ set2)

# symmetric_difference_update()
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.symmetric_difference_update(set2)
print("symmetric_difference_update():", set1)
````

# Matplotlib Basics

Matplotlib is a Python library used to create charts, graphs, and data visualizations.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Plotting X and Y Points

The `plot()` function is used to plot points and connect them with lines.

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y)
plt.show()
```

# 2. Plotting Multiple Points

Both arrays must have the same number of values.


```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)
plt.show()
```

# 3. Markers

Markers highlight individual data points.

### Common Markers

| Marker | Shape |
|--------|-------|
| `'o'` | Circle |
| `'*'` | Star |
| `'.'` | Point |
| `'x'` | X |
| `'+'` | Plus |
| `'s'` | Square |
| `'D'` | Diamond |
| `'^'` | Triangle Up |
| `'v'` | Triangle Down |

---

```python
y = np.array([3, 8, 1, 10])

plt.plot(y, marker="o")
plt.show()
```

# 4. Marker Size

Use `markersize` or `ms`.

```python
plt.plot(y, marker="o", ms=15)
plt.show()
```

# 5. Marker Color

### Edge Color


```python
plt.plot(y, marker="o", mec="red")
plt.show()
```

### Face Color

```python
plt.plot(y, marker="o", mfc="green")
plt.show()
```

### Edge and Face Color

```python
plt.plot(y, marker="o", mec="blue", mfc="blue")
plt.show()
```

# 6. Marker Colors
You can use:

- Color names (`"red"`, `"blue"`)
- Short names (`"r"`, `"g"`)
- Hex codes (`"#4CAF50"`)

```python
plt.plot(y, marker="o", mfc="#4CAF50")
plt.show()
```

# 7. Line Style

Change the appearance of the line.
### Line Styles

| Style | Short Form |
|-------|------------|
| solid | `-` |
| dotted | `:` |
| dashed | `--` |
| dashdot | `-.` |

---

```python
plt.plot(y, linestyle="dashed")
plt.show()
```

### Short Form

```python
plt.plot(y, ls=":")
plt.show()
```

# 8. Line Color

```python
plt.plot(y, color="red")
plt.show()
```

```python
plt.plot(y, c="#4CAF50")
plt.show()
```

# 9. Line Width

```python
plt.plot(y, linewidth=4)
plt.show()
```

# 10. Multiple Lines

Plot multiple datasets on the same graph.

```python
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```


# 11. Labels

Add labels to the x-axis and y-axis.

```python
plt.plot(y)

plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

# 12. Title

Add a title to the graph.

```python
plt.plot(y)

plt.title("Monthly Sales")

plt.show()
```


# 13. Label Color

```python
plt.xlabel("Months", color="red")
plt.ylabel("Sales", color="blue")

plt.plot(y)
plt.show()
```

# 14. Label Size

```python
plt.xlabel("Months", fontsize=14)
plt.ylabel("Sales", fontsize=14)

plt.plot(y)
plt.show()
```

# 15. Label Alignment
Possible values:

- `"left"`
- `"center"`
- `"right"`


```python
plt.title("Sales Report", loc="left")

plt.plot(y)
plt.show()
```

# 16. Font Style

```python
plt.title(
    "Sales Report",
    fontstyle="italic",
    fontweight="bold"
)

plt.plot(y)
plt.show()
```

# 17. Grid Lines
Display both horizontal and vertical grid lines.

```python
plt.plot(y)

plt.grid()

plt.show()
```

# 18. Horizontal Grid Only

```python
plt.plot(y)

plt.grid(axis="y")

plt.show()
```

# 19. Vertical Grid Only

```python
plt.plot(y)

plt.grid(axis="x")

plt.show()
```

# 20. Grid Color

```python
plt.plot(y)

plt.grid(color="red")

plt.show()
```

# 21. Grid Line Style

```python
plt.plot(y)

plt.grid(
    linestyle="--",
    linewidth=1
)

plt.show()
```

# 22. Subplots
Display multiple graphs in one figure.
`subplot(rows, columns, position)`

```python
x = [1,2,3]
y1 = [2,4,6]
y2 = [1,3,5]

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

# 23. Data Point Color

```python
plt.scatter(
    [1,2,3,4],
    [5,7,4,8],
    color="green"
)

plt.show()
```

# 24. Different Color for Each Point
Pass a list of colors.

```python
x=[1,2,3,4]
y=[5,7,4,8]

colors=["red","blue","green","orange"]

plt.scatter(x,y,c=colors)

plt.show()
```

# 25. Bar Chart
Used to compare different categories.

```python
students=["A","B","C","D"]
marks=[80,90,70,85]

plt.bar(students,marks)

plt.show()
```

### Horizontal Bar Chart

```python
plt.barh(students,marks)

plt.show()
```

# 26. Histogram

Shows the distribution (frequency) of numerical data.

```python
data=np.random.normal(170,10,100)

plt.hist(data)

plt.show()
```

# 27. Pie Chart
Shows each category as a percentage of the whole.
# Common Colors

| Short | Color |
|------|-------|
| `r` | Red |
| `g` | Green |
| `b` | Blue |
| `c` | Cyan |
| `m` | Magenta |
| `y` | Yellow |
| `k` | Black |
| `w` | White |

---

```python
marks=[40,30,20,10]
subjects=["Math","English","Science","CS"]

plt.pie(
    marks,
    labels=subjects,
    autopct="%1.1f%%"
)

plt.show()
```

# Most Frequent Value Imputation (Mode Imputation)

**Definition**

Most Frequent Value Imputation is a technique in which missing values are replaced with the **most frequently occurring value (mode)** of that column.

Although it can be applied to numerical data, it is mainly used for **categorical features**.

### When to Use
- When the feature is categorical.
- When the percentage of missing values is small.
- When the missing values are assumed to be random.

### Advantages
- Simple and easy to implement.
- Preserves the number of observations.

### Disadvantages
- Can change the distribution of the data.
- May introduce bias if many values are missing.

---

# Missing Category Imputation

Instead of replacing missing values with the mode, we can create a **new category** such as `"Missing"`.

This allows the model to learn that the value was originally missing instead of assuming it belongs to an existing category.

# Apply Most Frequent Value Imputation Using Scikit-learn

Split the dataset into training and testing sets before applying the imputer.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    df.drop(columns=['SalePrice']),
    df['SalePrice'],
    test_size=0.2
)
```

# Create the Imputer

Create a `SimpleImputer` that replaces missing values with the most frequent value.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
```

# Fit and Transform the Data

Learn the most frequent value from the training data and use it to replace missing values in both the training and testing datasets.

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

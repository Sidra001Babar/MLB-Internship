# Missing Category Imputation

**Definition**

Missing Category Imputation is a technique in which all missing values are replaced with a **new category**, such as **"Missing"**.

Instead of filling missing values with the most frequent category, this technique creates a separate category so that the model knows the value was originally missing.

### When to Use
- For **categorical features**.
- When missing values may contain useful information.
- When the data is **Not Missing At Random (NMAR)**.

### Advantages
- Very simple to implement.
- No data is lost.
- Preserves information that a value was missing.

### Disadvantages
- Introduces a new category that did not originally exist.
- May not improve model performance if missing values are completely random.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])   # load only required columns
df.head()
```

# Perform Missing Category Imputation Using Pandas

Replace all missing values in the **GarageQual** column with a new category called **"Missing"** using the Pandas `fillna()` method.

```python
df['GarageQual'].fillna('Missing', inplace=True)
```
